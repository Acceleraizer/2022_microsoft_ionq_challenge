```python
sep_cl['year'] = sep_cl["Submit Date Time"].dt.year
sep_cl['month'] = sep_cl["Submit Date Time"].dt.month
sep_cl['day'] = sep_cl["Submit Date Time"].dt.day
```

```python
#for each day, have # of events in xx department

myDict = {aug_cl["day"].iloc[0]: {aug_cl['Customer Department'].iloc[0]: 1}}

for index, row in aug_cl.iterrows():
    if row['day'] not in myDict:
        myDict[row['day']] = {}
    print('row date', row['day'])
    if row['Customer Department'] not in myDict[row['day']]:
        myDict[row['day']][row['Customer Department']] = 0
    myDict[row['day']][row['Customer Department']] = myDict[row['day']][row['Customer Department']] + 1
    
#need to order dates tho...
```
